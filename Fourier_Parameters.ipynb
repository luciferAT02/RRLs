# Code snippet to obtain the Fourier Parameters 

This part of the code imports all the required modules for analysis

```python
import matplotlib.pyplot as plt 
from astropy.io import fits
import numpy as np
import math
import pandas as pd
import Fourier_Fit as fft
import phase_magnitude as pm
import csv
from matplotlib.backends.backend_pdf import PdfPages
from Loess import Loess
from sklearn.neighbors import KernelDensity
```

This part of the code deals with the extraction of the photometric data, converting the available G-band magnitude to V-band magnitude, phasing the light curve and then determining the Fourier parameters to study about the shape of the curve.

```python
df=pd.read_csv('merged_data.txt') # Reads the files with gaiaid, period, feh, feh error and extinction
period=df['period']
gaiaid=df['gaiaid']
feh=df['feh']
feh_e=df['feh_e']
extinction=df['AV_SandF']

def write_fourier_params(gaia_id, period,params, eparams, filename='efourier_parameters_n4_1.txt', param_names=['m0','A1', 'A2', 'A3', 'A4', 'p1', 'p2', 'p3', 'p4'], eparam_names=['em0','eA1', 'eA2', 'eA3', 'eA4', 'ep1', 'ep2', 'ep3', 'ep4']):  # Add placeholder parameter names
    """
    Writes Fourier parameters to a CSV file.
    
    Args:
    gaia_id (str): Gaia ID of the star.
    period (float): Period of the light curve.
    params (list or np.ndarray): List or NumPy array containing Fourier parameters.
    filename (str, optional): Name of the output CSV file. Defaults to 'fourier_parameters.csv'.
    param_names (list, optional): List of parameter names (optional, if order matters). Defaults to ['p1', 'p2', ...].
    """
    with open(filename, 'a', newline='') as csvfile:
        writer = csv.writer(csvfile)
        
        # Write header row on first iteration only
        
        if csvfile.tell() == 0:  # Check if file is empty
            writer.writerow([str('gaiaid'), 'period'] + (param_names if param_names else [f'param_{i+1}' for i in range(len(params))])+ (eparam_names if eparam_names else [f'eparam_{i+1}' for i in range(len(eparams))]))  # Use placeholder names if not provided
        writer.writerow([gaia_id, period] + list(params)+list(eparams))


with PdfPages("efourier_light_curves_n4_1.pdf") as pdf: 
    for i in range(len(period)):
        phase, mag, period[i]=pm.vband(period[i],gaiaid[i])
        mag=mag-extinction[i]
        phase=np.array(phase)
        
        # 2 sigma clipping:
        
        lower_bound = np.mean(mag) - 2*np.std(mag)
        upper_bound = np.mean(mag) + 2*np.std(mag)
        mask = (mag >= lower_bound) & (mag <= upper_bound)
        mag = mag[mask]
        phase = phase[mask]
        
        FS=fft.FourierSeries()
        params, e_params = FS.fit(phase, mag, n=4)
        _x = np.linspace(0, 1, 200)  # Create a finer grid of phase values
        _x = np.array(_x)
        y_fit,yerr_fit = FS.generate(_x, params,)
        fig=plt.figure()
        plt.xlabel("Phase")
        plt.ylabel("Magnitude")
        plt.title("Fourier-Fitted Light Curve of "+str(gaiaid[i]))
        plt.scatter(phase, mag, label='Phased Data')
        plt.plot(_x, y_fit, label='Fitted Curve')
        plt.show()
        write_fourier_params(gaiaid[i], period[i], params,e_params)
        fig.savefig(pdf,format="pdf")
        plt.close(fig)
```
